```python
def menu_visual_agenda():
    # Função para exibir o menu visual
    print("\n[1] Inserir novo contato")
    print("[2] Consultar contato por nome")
    print("[3] Remover contato por nome")
    print("[4] Alterar contato")
    print("[5] Incluir múltiplos contatos")
    print("[6] Gerar relatório")
    print("[0] Sair")
    return int(input("Escolha uma opção: "))

def abrir_arquivo():
    # Função para abrir o arquivo.
    try:
        with open("FileNames.txt", "r") as arquivo:
            linhas = arquivo.readlines()
        return linhas
    except FileNotFoundError:
        print("Arquivo não encontrado.")
        return []
    except IOError as e:
        print(f"Erro ao abrir o arquivo: {e}")
        return []

def salvar_agenda(agenda):
    try:
        with open("FileNames.txt", "w") as arquivo:
            for contato, dados in agenda.items():
                telefone = dados.get("Telefone", "")
                email = dados.get("E-mail", "")
                twitter = dados.get("Twitter", "")
                instagram = dados.get("Instagram", "")
                arquivo.write(f"{correcao(contato)};{correcao(telefone)};{correcao(email)};{correcao(twitter)};{correcao(instagram)}\n")
        print("Agenda salva com sucesso!")
    except IOError as e:
        print(f"Erro ao escrever no arquivo: {e}")
        
def carregar_agenda():
    linhas = abrir_arquivo()
    agenda = {}
    for contato in linhas:
        campos = contato.strip().split(";")
        contato = campos[0]
        telefone = campos[1]
        email = campos[2]
        twitter = campos[3]
        instagram = campos[4]
        agenda[contato] = {"Telefone": telefone, "E-mail": email, "Twitter": twitter, "Instagram": instagram}
    return agenda
```

```python
def inserir(agenda):
    # Função que cadastra contatos
    try:
        nome = str(input("\nQual nome?\n|--> ").strip())
        telefone = str(input("Qual número de telefone?\n|--> 61-9").strip())
        email = str(input("Qual E-mail?\n|--> ").strip())
        twitter = str(input("Qual twitter?\n|--> ").strip())
        instagram = str(input("Qual instagram?\n|--> ").strip())
        agenda[correcao(nome)] = {"Telefone": correcao(telefone), "E-mail": correcao(email), "Twitter": correcao(twitter), "Instagram": correcao(instagram)}
    except ValueError:
        print('\nNo telefone insira apenas números')
    else:
        print("Contato inserido com sucesso!")
        return agenda

def correcao(x):
    # Função para tratar campos vazios, substituindo por <desconhecido>
    return x.strip() if x.strip() else '<desconhecido>'

def incluir_multiplos(agenda, n):
    for _ in range(n):
        inserir(agenda)

def consultar(agenda):
    # Função para consultar um contato por nome.
    nome = input("Digite o nome do contato a ser consultado: ")
    if nome in agenda:
        print("Nome:", nome)
        for key, value in agenda[nome].items():
            print(key + ":", value)
    else:
        print("Contato não encontrado.")

def remover(agenda):
    # Função para remover um contato por nome.
    nome = input("Digite o nome do contato a ser removido: ")
    if nome in agenda:
        del agenda[nome]
        print(f"Contato {nome} removido com sucesso!")
    else:
        print("Contato não encontrado.")

        
def alterar(agenda):
    # Função para alterar um contato
    nome = input("Digite o nome do contato a ser alterado: ")
    if nome in agenda:
        print("Digite as novas informações:")
        telefone = input("Telefone: ")
        email = input("E-mail: ")
        twitter = input("Twitter: ")
        instagram = input("Instagram: ")
        agenda[nome] = {"Telefone": telefone, "E-mail": email, "Twitter": twitter, "Instagram": instagram}
        print("Contato alterado com sucesso!")
    else:
        print("Contato não encontrado.")

def gerar_relatorio(agenda):
    if not agenda:
        print("A agenda está vazia. Não há contatos cadastrados.")
        return

    print("{:<5} {:<15} {:<15} {:<20} {:<15} {:<15}".format("Nro", "Nome", "Telefone", "E-mail", "Twitter", "Instagram"))
    print("=" * 105)

    for n, (nome, dados) in enumerate(agenda.items(), 1):
        telefone = correcao(dados["Telefone"])
        email = correcao(dados["E-mail"])
        twitter = correcao(dados["Twitter"])
        instagram = correcao(dados["Instagram"])
        print("{:<5} {:<15} {:<15} {:<20} {:<15} {:<15}".format(n, nome, telefone, email, twitter, instagram))
```

```python
def principal():
    agenda = carregar_agenda()
    while True:
        opcao = menu_visual_agenda()
        if opcao == 0:
            salvar_agenda(agenda)
            print("Agenda finalizada.")
            break
        elif opcao == 1:
            agenda = inserir(agenda)
            salvar_agenda(agenda)
        elif opcao == 2:
            consultar(agenda)
        elif opcao == 3:
            remover(agenda)
            salvar_agenda(agenda)
        elif opcao == 4:
            alterar(agenda)
            salvar_agenda(agenda)
        elif opcao == 5:
            n = int(input("Digite o número de contatos que deseja incluir: "))
            incluir_multiplos(agenda, n)
            salvar_agenda(agenda)
        elif opcao == 6:
            gerar_relatorio(agenda)
        else:
            print("Opção inválida. Tente novamente.\n")
```
